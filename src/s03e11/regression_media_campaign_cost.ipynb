# Kaggle competition 3.11: Regression with media campaign cost dataset

```python
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
train.head()
```

```python
X = train.drop('cost', axis=1)
y = train['cost']
model = linear_model.LinearRegression()
model.fit(X, y)
predictions = model.predict(test)
results = pd.DataFrame({'id': test['id'], 'cost': predictions})
results.to_csv('result1.csv', index=False)
```

## Analyzing the features

```python
train_features = train.copy()
train_features.pop('id')
train_features.corr()
```

salad_bar and prepared_food are almost identical in correlation. Therefore, we drop the first one.

```python
train_features.pop('salad_bar')
train_features.corr()
costs = train_features.pop('cost')
reg = linear_model.LinearRegression()
reg.fit(train_features, costs)
```

```python
test2 = test.copy()
test2.pop('id')
test2.pop('salad_bar')
```

```python
result = reg.predict(test2)
result2 = pd.concat([test2, pd.DataFrame({'cost': result[:]})], axis=1)
result2
```
